```python
import numpy as np
from matplotlib import pyplot as plt
import glob
```

```python
datalist  = '/rsrch1/ip/jacctor/smallnet/tumorhcc-2D/alldata/data/'
imglist = glob.glob(datalist+'img/volume-023*.npy')
seglist = glob.glob(datalist+'seg/segmentation-023*.npy')
print(len(imglist))
print(len(seglist))
```

```python
def show_figs(img, seg):
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(img[...,0])
    plt.subplot(122)
    plt.imshow(seg[...,0])
    plt.show()
```

```python
ilist = [i for i in imglist]
ilist.sort()

slist = [s for s in seglist]
slist.sort()
```

```python
img = np.load('img/volume-1-10.npy')
seg = np.load('seg/segmentation-1-10.npy')
print(img.dtype)
print(seg.dtype)
```

```python
for i in range(0,len(ilist),20):
    print(i,':')
    print(ilist[i][len(datalist):])
    print(slist[i][len(datalist):])
    img = np.load(ilist[i])
    seg = np.load(slist[i])
    show_figs(img, seg)
```

```python
liver_size = [None]*len(slist)
tumor_size = [None]*len(slist)
for idx, s in enumerate(slist):
    seg = np.load(s)
    liver = np.sum(seg > 0)
    tumor = np.sum(seg > 1)
    liver_size[idx] = liver
    tumor_size[idx] = tumor
#     print(s[len(datalist):], ':\t', liver, tumor)
```

```python
nonzero_tumors = [t for t in tumor_size if t > 0]
print('size     0 tumors', len([t for t in tumor_size if t ==  0]))
print('size <   5 tumors', len([t for t in tumor_size if 0 <  t <   5]))
print('size <  10 tumors', len([t for t in tumor_size if 5 <= t <  10]))
print('size <  50 tumors', len([t for t in tumor_size if 10<= t <  50]))
print('size >= 50 tumors', len([t for t in tumor_size if t >= 50]))
print()
print('size     0 livers', len([t for t in liver_size if t ==  0]))
print('size <   5 livers', len([t for t in liver_size if 0 <  t <   5]))
print('size <  10 livers', len([t for t in liver_size if 5 <= t <  10]))
print('size <  50 livers', len([t for t in liver_size if 10<= t <  50]))
print('size >= 50 livers', len([t for t in liver_size if t >= 50]))


plt.figure(figsize=(10,5))
plt.subplot(121)
plt.hist(liver_size, bins=100)
plt.subplot(122)
plt.hist(nonzero_tumors, bins=100)
plt.show()
```
